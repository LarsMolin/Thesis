```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("main.csv", sep=';')
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S%z', utc=True)
df.set_index('Date', inplace=True)
df.index = df.index.tz_convert('Europe/Amsterdam')
```

```python
df['Final_Price'] = np.where(~df['Short_Price'].isna(), df['Short_Price'],
                    np.where(~df['Long_Price'].isna(), df['Long_Price'], df['Mid_Price']))
df.drop(columns=['Long_Price', 'Short_Price'], inplace=True)
```

```python
df.drop(columns=['BE to NL (from Belgium)', 'DE to NL (from Germany)',
       'DK to NL (from Denmark)', 'NO to NL (from Norway)',
       'NL to BE (to Belgium)', 'NL to DE (to Germany)',
       'NL to DK (to Denmark)', 'NL to NO (to Norway)','Belgium Intraday',
       'Denmark Intraday', 'Germany Intraday', 'Great Britain Intraday',
       'Norway Intraday'], inplace=True)
```

```python
df['Dispatch'] = df['Upward_Dispatch'] - df['Downward_Dispatch']
df['IGCC_Dispatch'] = df['IGCCCONTRIBUTION_UP'] - df['IGCCCONTRIBUTION_DOWN']
```

```python
df.drop(columns=['Upward_Dispatch','Downward_Dispatch', 'IGCCCONTRIBUTION_UP', 'IGCCCONTRIBUTION_DOWN'], inplace=True)
```

```python
for i in df.columns:
    print(i,sum(df[i].isna()))

```

```python
df['MinuteOfDay'] = df.index.hour * 60 + df.index.minute
df['HourOfDay'] = df.index.hour
df['Imbalance_Price -2H'] = df['Final_Price'].shift(periods=-2, freq='1H')
df['IGCC_Dispatch -2H'] = df['IGCC_Dispatch'].shift(periods=-2, freq='1H')
df['Dispatch -2H'] = df['Dispatch'].shift(periods=-2, freq='1H')
df['Imbalance_Price -5m'] = df['Final_Price'].shift(periods=-5, freq='1T')
df['IGCC_Dispatch -5m'] = df['IGCC_Dispatch'].shift(periods=-5, freq='1T')
df['Dispatch -5m'] = df['Dispatch'].shift(periods=-2, freq='5T')
df['WIND OUTTURN'] = df['WIND ONSHORE OUTTURN'] + df['WIND OFFSHORE OUTTURN']
df['NETTO SOLAR'] = df['SOLAR OUTTURN'] - df['DAY AHEAD FORECAST (ENAPPSYS)']
df['NETTO WIND'] = df['FORECAST D-1 (ENAPPSYS)'] - df['WIND OUTTURN']
df['Final_Price +2H'] = df['Final_Price'].shift(periods=2, freq='1H')
```

```python
df = df["2021-12-31 23:00:00+01:00":"2022-12-31 21:59:00+01:00"]
```

```python
from sklearn.impute import KNNImputer
df = df.loc[:, [
                'DAY AHEAD FORECAST (ENAPPSYS)',
                'SOLAR FORECAST UNADJUSTED (ENAPPSYS)',
                'WIND FORECAST UNADJUSTED (ENAPPSYS)',
                'FORECAST D-1 (ENAPPSYS)',
                'NETTO SOLAR',
                'NETTO WIND',
                'INFERRED INSTALLED CAPACITY (ENAPPSYS)_x',
                'SOLAR OUTTURN',
                'WIND OUTTURN',
                'INFERRED INSTALLED CAPACITY (ENAPPSYS)_y',
                'DA Scheduled Flow',
                'Scheduled Flow',
                'Final_Price',
                'Final_Price +2H', 
                'Dispatch', 
                'IGCC_Dispatch',
                'Dispatch -2H', 
                'HourOfDay',
                'MinuteOfDay',
                'IGCC_Dispatch -2H',
                'Imbalance_Price -2H',
                'Dispatch -5m',
                'IGCC_Dispatch -5m',
                'Imbalance_Price -5m',
                'minmax',
                'min600',
                'min300',
                'min100',
                'minmin',
                'posmin',
                'pos100',
                'pos300',
                'pos600',
                'posmax',
                'NETHERLANDS (NL)'
                ]]

original_index = df.index

imputer = KNNImputer(n_neighbors=5)
df_imputed = imputer.fit_transform(df)

df = pd.DataFrame(data=df_imputed, columns=df.columns, index=original_index)
```
